```python
import pandas as pd

# Takes the lyrics of each song and concatenates them into a single text file, delimited by a single whitespace.
```

```python
songs = pd.read_csv("drake-songs-spotify.csv", index_col=False)
songs
```

```python
lyrics = songs[["lyrics"]]


texts = []
for i in range(lyrics.size):
    texts.append(lyrics["lyrics"].iloc[i])

trainable_data = " ".join(texts)
print(trainable_data)
```

```python
with open("../models/char_rnn/data/drake_data.txt", "w") as text_file:
    text_file.write(trainable_data)
```
